Q1.What is unsupervised learning in the context of machine learning?

Answer: Unsupervised learning is a type of machine learning where the model learns patterns from data without labeled outputs. It tries to find hidden structures or relationships in the data, such as grouping (clustering) or association.


Q2.How does K-Means clustering algorithm work?

Answer: K-Means works by:

1.Selecting K initial cluster centroids

2.Assigning each data point to the nearest centroid

3.Updating centroids as the mean of assigned points

4.Repeating steps until centroids no longer change

Q3.Explain the concept of a dendrogram in hierarchical clustering?

Answer: A dendrogram is a tree-like diagram that shows how clusters are formed in hierarchical clustering. It represents the merging or splitting of clusters at different distance levels.

Q4.What is the main difference between K-Means and Hierarchical Clustering?

Answer: K-Means requires the number of clusters (K) in advance, while hierarchical clustering does not. K-Means is faster, while hierarchical clustering provides a tree structure of clusters.

Q5.What are the advantages of DBSCAN over K-Means?

Answer:

*Can find clusters of arbitrary shapes

*Does not require number of clusters beforehand

*Handles noise and outliers well

Q6.When would you use Silhouette Score in clustering?

Answer: Silhouette Score is used to evaluate clustering quality. It helps determine how well-separated and compact the clusters are.

Q7.What are the limitations of Hierarchical Clustering?

Answer:

*Computationally expensive for large datasets

*Difficult to undo merges or splits

*Sensitive to noise and outliers

Q8.Why is feature scaling important in clustering algorithms like K-Means?

Answer: Feature scaling ensures that all features contribute equally. Without scaling, features with larger values dominate distance calculations.

Q9.How does DBSCAN identify noise points?

Answer: DBSCAN labels points as noise if they do not have enough neighboring points within a specified radius (epsilon).

Q10.Define inertia in the context of K-Means?

Answer: Inertia is the sum of squared distances between data points and their nearest cluster centroid. It measures how tightly the clusters are formed.

Q11.What is the elbow method in K-Means clustering?

Answer: The elbow method helps find the optimal number of clusters by plotting inertia vs. number of clusters and identifying the “elbow point” where the decrease slows down.

Q12.Describe the concept of "density" in DBSCAN?

Answer: Density refers to how closely data points are packed together. DBSCAN groups points that are closely packed and separates sparse regions as noise.

Q13.Can hierarchical clustering be used on categorical data?

Answer: Yes, hierarchical clustering can be used with categorical data if an appropriate distance measure (like Hamming distance) is used.

Q14.What does a negative Silhouette Score indicate?

Answer: A negative Silhouette Score indicates that data points may be assigned to the wrong cluster and clusters are overlapping.

Q15.Explain the term "linkage criteria" in hierarchical clustering?

Answer: Linkage criteria determine how the distance between clusters is calculated (e.g., single, complete, average linkage).

Q16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Answer: K-Means assumes clusters are spherical and equally sized. It struggles when clusters have different shapes, sizes, or densities.

Q17.What are the core parameters in DBSCAN, and how do they influence clustering?

Answer:

*Epsilon (ε): Defines neighborhood radius

*MinPts: Minimum number of points to form a cluster

 These parameters control cluster formation and noise detection.

Q18.How does K-Means++ improve upon standard K-Means initialization?

Answer: K-Means++ selects initial centroids more carefully to improve convergence speed and avoid poor clustering results.

Q19.What is agglomerative clustering

Answer: Agglomerative clustering is a type of hierarchical clustering that starts with individual data points and merges them step-by-step into clusters.

Q20.What makes Silhouette Score a better metric than just inertia for model evaluation?

Answer: Silhouette Score considers both cluster cohesion and separation, while inertia only measures compactness. Hence, it provides a more balanced evaluation.

```python
#Q29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Store silhouette scores
k_values = [2, 3, 4, 5]
scores = []

# Compute silhouette score for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

# Print scores
for k, score in zip(k_values, scores):
    print(f"k = {k}, Silhouette Score = {score:.4f}")

# Plot bar chart
plt.bar(k_values, scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different k Values")
plt.show()
```

```python
#Q31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data (overlapping clusters)
X, y = make_blobs(n_samples=300, 
                  centers=3, 
                  cluster_std=2.5,   # high std = overlapping
                  random_state=42)

# Step 2: Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 3: Create mesh grid for decision boundaries
h = 0.02  # step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict cluster for each point in mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3)

# Plot original data
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, edgecolor='k')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], 
            kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centroids')

plt.title("K-Means with Overlapping Clusters")
plt.legend()
plt.show()
```

```python
#Q34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 1: Load dataset
data = load_breast_cancer()
X = data.data

# Step 2: Standardize data
X_scaled = StandardScaler().fit_transform(X)

# Step 3: Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Step 4: Plot the results
plt.figure(figsize=(8,6))
plt.plot(k_values, inertia_values, marker='o')

plt.title("Inertia vs Number of Clusters (K-Means)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")

plt.xticks(k_values)
plt.grid()

plt.show()
```

```python
#Q44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples

import warnings
warnings.filterwarnings('ignore')

# Load Iris dataset
iris = load_iris()
X = iris.data

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Compute silhouette values for each sample
silhouette_vals = silhouette_samples(X, labels)

# Plot Silhouette Coefficient for each sample
plt.figure()

y_lower = 0
for i in range(3):
    ith_cluster_vals = silhouette_vals[labels == i]
    ith_cluster_vals.sort()
    size_cluster = ith_cluster_vals.shape[0]
    y_upper = y_lower + size_cluster

    plt.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_vals
    )

    y_lower = y_upper

plt.title("Silhouette Plot for Iris Dataset (KMeans)")
plt.xlabel("Silhouette Coefficient")
plt.ylabel("Samples")

plt.show()
```

```python
#Q46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

# Load Wine dataset
wine = load_wine()
X = wine.data

# Convert to DataFrame (first 4 features)
df = pd.DataFrame(X[:, :4], columns=wine.feature_names[:4])

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Pairplot visualization
sns.pairplot(df, hue='Cluster')

plt.show()
```

```python
#Q47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

# Generate noisy blobs
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

# Count clusters and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print("Number of clusters:", n_clusters)
print("Number of noise points:", n_noise)
```

```python
#Q48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

# Load Digits dataset
digits = load_digits()
X = digits.data

# Apply t-SNE (reduce to 2D)
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=10)
labels = agg.fit_predict(X_tsne)

# Plot clusters
plt.figure()
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels)

plt.title("Agglomerative Clustering on t-SNE Reduced Digits Dataset")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")

plt.show()
```
